```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:\Machine Learning and Data Science\gene_expression.csv')
df.head()
```

```python
sns.scatterplot(data = df,x = 'Gene One',y = 'Gene Two',hue = 'Cancer Present',alpha = 0.3)
plt.xlim(4,6.5)
plt.ylim(4,8)
```

```python
sns.pairplot(data = df,hue = 'Cancer Present')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop('Cancer Present',axis  = 1)
y = df['Cancer Present']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.35,random_state = 42)
```

```python
scaler = StandardScaler()
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.transform(x_test)        

#Since y_train is either 0 or 1 i.e it  doesn't need to scale
 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model =  KNeighborsClassifier(n_neighbors=1)
# help(KNeighborsClassifier)
model.fit(scaled_x_train,y_train)      
y_predicted = model.predict(scaled_x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_ = accuracy_score(y_test,y_predicted)
accuracy_
```
